```python
#Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('winequality-red.csv')
```

```python
df.head()
```

```python
#Assining X and y values

X = df.drop('quality',axis=1)
y = df[['quality']]

print(X.columns)
print(y.columns)
```

```python
X.head()
```

```python
y.head()
```

```python
df.describe()
```

```python
df.hist(bins=25,figsize=(10,10))
# display histogram
plt.show()
```

```python
plt.figure(figsize=[10,6])
# plot bar graph
plt.bar(df['quality'],df['pH'],color='#001AF5')
# label x-axis
plt.xlabel('quality')
#label y-axis
plt.ylabel('pH')
```

```python
# ploting heatmap
plt.figure(figsize=[19,10],facecolor= 'white')
sns.heatmap(df.corr(),annot=True)
```

```python
#Importing Model

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100) #n_estimators is the of trees that are produced in voting
model.fit(X, y)
model.predict([[7.4, 0.70, 0.00, 1,9, 0.076, 11.0, 34.0, 0.997, 3.51, 0.56]])
```

### I choose 0.2 test size, because it 1/4 ratio give us best accuracy of the model.

```python
#Model Implementation

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
predictions = model.predict(X_test)
predictions
```

```python
#Accuracy Test

score = model.score(X_test, y_test)
print("The Accuracy Score is: ", score)
```

```python
from sklearn import metrics

cm= metrics.confusion_matrix(y_test, predictions)
cm
```

```python
#Plotting
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square= True, cmap= 'Spectral');
plt.ylabel('Actual Output');
plt.xlabel('Predicted Output');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);
```
