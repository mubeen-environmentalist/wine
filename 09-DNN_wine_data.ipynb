```python
# Import Train Test Splitting
from sklearn.model_selection import train_test_split

#X values
X = X = df.drop('quality', axis=1)
#y value
y = df['quality']

#train test split
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.2, random_state=0) #random split into train and test 

#print the size of X and y (train and test)
print("X_train :", X_train.size)
print("X_test :", X_test.size)
print("y_train :", y_train.size)
print("X_test :", y_test.size)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
print("Transformed X_train :", X_train)
X_test = scaler.fit_transform(X_test)
print("Transformed X_test :", X_test)
```

```python
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense
from tensorflow.keras import layers, Input
from keras.layers import ReLU

#Model Explaination
model = Sequential()  #empty model ; Xavier Methodology for Initialization 

model.add(Dense(30, activation='relu'))  #1st hidden layer: 30 = number of units/neurons

#newly added hidden layer
model.add(Dense(20, activation='relu')) #add one at a time

# output layer of binary classification
model.add(Dense(1, activation='sigmoid'))
```

# Model is ready to build as a basis for performance evaluation

```python
#call back function
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    '''
    Halts the data after 60 percent accuracy

    Args:
      epoch (integer) - index of epoch (required but unused in the function defination below)
      logs (dist) - metric results from the training epoch
    '''

    # check accuracy
    if(logs.get('loss') < 0.15):

      #stop if threshhold meet
      print("\nLoss is lower than 0.15 so canceling training")
      self.model.stop_training = True



# Instantiate Class
callbacks = myCallback()
```

```python
#define the model
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

#Compile the model
model.compile(tf.optimizers.Adam(),
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'],
)
```

```python
history = model.fit(
    x=X_train,
    y=y_train,
    batch_size=15000,
    epochs=150,
    callbacks=[callbacks],
)
```

## The accuracy of model is 94%, this is very high accuracy.

```python
loss,accuracy = model.evaluate(X_test, y_test, verbose=False)
```

```python
#Evaluate the model on unseen data
model.evaluate(X_test, y_test)
```
